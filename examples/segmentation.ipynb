```python
# Get the model repo url
models_catalog = default_catalog.models.to_dataframe()
model_repo = models_catalog[models_catalog.name == "huggingface-segmentation"].url.item()
model_repo # Why not paste the repo link into your browser and see how it looks?
```

```python
facebook_model = load_pretrained_model(model_repo, allow_install="force", model='facebook_detr_resnet_50_panoptic')
#                                       branch="fe49a6ab2aaffddaa7524474f3e29ab46fb23d68")
nvidia_model = load_pretrained_model(model_repo, model='nvidia_segformer_b5_finetuned_ade_640_640')
microsoft_model = load_pretrained_model(model_repo, model='microsoft_beit_large_finetuned_ade_640_640')
```

```python
import cv2
from detectron2.utils.visualizer import Visualizer, ColorMode
from PIL import Image
import numpy as np
```

```python
pillow_image = Image.fromarray(sample_image.to_numpy(), 'RGB')
im = np.array(pillow_image)
im = im[:, :, ::-1].copy() # Convert RGB to BGR 
```

```python
v = Visualizer(im[:, :, ::-1], scale=1.5, instance_mode=ColorMode.IMAGE_BW)   # remove the colors of unsegmented pixels
v = v.draw_instance_predictions(ms_logits["instances"].to("cpu"))
image = cv2.cvtColor(v.get_image()[:, :, :], cv2.COLOR_BGR2RGB)

plot_predictions = plt.figure(figsize=(15,15))
plt.imshow(Image.fromarray(image))
plt.title('Predictions',fontsize='xx-large')
plt.axis('off')
plt.show()
```
